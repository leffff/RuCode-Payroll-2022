```python
import pickle
from sklearn.decomposition import TruncatedSVD
import numpy as np
import os
from tqdm import tqdm
```

```python
from google.colab import drive
drive.mount('/content/drive')
TEXT_EMBEDINGS = "/content/drive/MyDrive/payroll/"
embedding_names = os.listdir(TEXT_EMBEDINGS + "text_embeddings")
embedding_names
```

```python
def get_svd_dict(unique_embeddings):
    keys = list(unique_embeddings.keys())
    svd = TruncatedSVD(n_components=8, n_iter=30, random_state=42)
    print("SVD Started")
    decreased_dims = svd.fit_transform(np.vstack(list(unique_embeddings.values()))) 
    print("SVD Finished")
    new_dict = dict()
    for i in tqdm(range(len(keys))):
        new_dict[keys[i]] = decreased_dims[i]
    return new_dict
```

```python
for filename in tqdm(embedding_names):
    with open(f"{TEXT_EMBEDINGS}text_embeddings/{filename}", "rb") as fin:
        unique_embeddings = pickle.load(fin)
    svd_embeddings = get_svd_dict(unique_embeddings)
        # print(svd_embeddings)
    with open(f"{TEXT_EMBEDINGS}svd_text_embeddings/{filename}", "wb") as fout:
        unique_embeddings = pickle.dump(svd_embeddings, fout)
```
